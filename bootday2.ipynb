````python
n = 5
row = 0
while row < n:
 col  = 0
 while col <= row:
  print("*",end=" ")
  col = col + 1
 print() 
 row = row + 1
````

````python
n = 5
row = 0
while row < n:
 col  = 0
 while col < (n-row):
  print("*",end=" ")
  col = col + 1
 print() 
 row = row + 1
````
